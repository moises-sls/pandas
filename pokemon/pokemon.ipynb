### Alterando os dados

```python
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df["Sp. Def"] + df['Speed']

df.head()
```

```python
df['Total'] = df.iloc[:, 4:10].sum(axis=1) # pega todas as linhas [:, ] e as colunas 'HP' 'Attack' 'Defense' ... até a coluna 9 (a 10 é excluida)
df.head()
```

```python
# reordenando as colunas para deixar a coluna 'Total' no começo
cols = list(df.columns.values)  # adicionando todos os nomes das colunas do dataframe em uma lista
df = df[cols[0:4] + [cols[-1]] + cols[4:12]] # seleciono as primeiras 3 colunas, depois a ultima(Total) e por fim as ultimas colunas restantes
df
```

### Filtrando os dados

```python
new_df = df.loc[ (df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df['HP'] > 50)]
new_df
```

```python
# como dropar todos os pokemons 'Mega':
df.loc[df['Name'].str.contains('Mega')] # pegamos o parâmetro string dos nomes dos pokemons e vemos se ele contem a palavra 'Mega'

# para selecionar todos os pokemons que NÃO são 'Mega':
df.loc[~df['Name'].str.contains('Mega')]
```

```python
# Utilizando regex
import re

df.loc[df['Type 1'].str.contains('fire|grass',flags=re.I, regex=True)]  # o operador | significa "ou" no regex
# o argumento "flags= re.I", indica quais observações devem ser utilizadas, nesse caso dizemos para 
# ignorar a capitalização, ou seja pegar palavras maísculas ou minúsculas
```

```python
df.loc[df['Name'].str.contains('^pi[a-z]*', flags=re.I, regex=True)] # ^ indica que deve analisar o começo da palavra
# "pi" indica que a palavra deve conter pi, e após pi pode conter de a-z, zero ou mais.
```

### Estatísticas Agregadas (Groupby)

```python
df.head()
```

```python
# a média de vida de um pokemon varia dependendo do seu tipo?

df.groupby(['Type 1']).mean(numeric_only=True).sort_values('HP', ascending=False)
 #numeric_only=True porque existem valores NA
```

```python
# qual tipo de pokemon, em média, tem a maior defesa?

df.groupby('Type 1').mean(numeric_only=True).sort_values('Defense',ascending=False)
```

```python
# qual tipo de pokemon, em média, possui o maior ataque?

df.groupby('Type 1').mean(numeric_only=True).sort_values('Attack', ascending=False)
```

```python
# qual tipo de pokemon, em média, tem a maior vida? 

df.groupby('Type 1').mean(numeric_only=True).sort_values('HP', ascending=False)

# podemos utilizar .sum() .median() .count() etc 
```

```python
# contando quantos pokemons de cada tipo existem no banco de dados

df['Count'] = 1

df.groupby('Type 1')['Count'].count()
```

```python
# podemos contar cada subconjunto:

df.groupby(['Type 1', 'Type 2'])['Count'].count()
```
